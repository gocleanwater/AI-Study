### 인덱싱과 슬라이싱

Python 인덱스는 **0부터 시작**합니다.

```python
ex_list[start:end:step]
```

- start: 시작 인덱스
- end: 끝 인덱스, 단 포함되지 않음
- step: 몇 칸씩 이동할지 (양수이면 양수만큼 뛰어넘기, 음수이면 방향 거꾸로해서 뛰어넘기)

`끝` 위치의 원소는 포함하지 않습니다.


````python
values = [10, 20, 30, 40, 50]

print(values[0])  # 10
print(values[1:4])  # [20, 30, 40]
print(values[:3])  # [10, 20, 30]
print(values[2:])  # [30, 40, 50]
print(values[::2])  # step이 2이므로 2칸씩 뛰어넘는다 → [10, 30, 50]
print(values[::-1])  # step이 -1이므로 방향이 거꾸로이다. → [50, 40, 30, 20, 10]
print(values[::-2])  # step이 -2이므로 뒤에서부터 -2칸씩 뛰어넘는다. → [50, 30, 10]
````

## 2. 함수

함수는 반복해서 사용할 코드를 하나로 묶은 것입니다.

```python
def 함수이름(매개변수):
    실행할 코드
    return 반환값
```

### 매개변수와 인자

```python
def add(a, b):
    return a + b

add(2, 3)
```

- `a`, `b`: 함수 정의에 적힌 **매개변수(parameter)**
- `2`, `3`: 함수를 호출할 때 전달한 **인자(argument)**


````python
# CNN shape 계산 공식 함수
def calculate_output_size(input_size, kernel_size, stride=1, padding=0):
    output_size = (input_size + 2 * padding - kernel_size) // stride + 1
    return output_size


result = calculate_output_size(
    input_size=32,
    kernel_size=3,
    stride=1,
    padding=1,
)

print(result)
````

### 기본값이 있는 매개변수

위 함수에서 `stride=1`, `padding=0`은 기본값입니다.

따라서 값을 전달하지 않으면 기본값이 사용됩니다.


````python
print(calculate_output_size(32, 3))
print(calculate_output_size(32, 3, stride=2))
print(calculate_output_size(32, 3, stride=2, padding=1))
````

## 3. 클래스, `self`, `__init__`

PyTorch 모델은 보통 클래스로 작성합니다.

```python
class MyModel:
    def __init__(self, input_size):
        self.input_size = input_size
```

### `self`

`self`는 **현재 생성된 객체 자기 자신**을 가리킵니다. (`Java`로 따지면 생성자 함수격이다.)

### `__init__`

객체를 생성할 때 자동으로 실행되는 초기화 메서드입니다.

### `input_size`

`input_size`는 객체를 만들 때 외부에서 전달받는 값입니다.

```python
model = MyModel(784)
```

이때 `784`가 `input_size`로 전달됩니다.


````python
class SimpleModel:
    def __init__(self, input_size, output_size):
        self.input_size = input_size
        self.output_size = output_size

    def describe(self):
        print(f"입력 크기: {self.input_size}")
        print(f"출력 크기: {self.output_size}")


model = SimpleModel(input_size=784, output_size=10)

model.describe()
````

### 객체마다 값이 따로 저장됩니다

`self.input_size`는 각 객체 내부에 저장된 변수입니다.


````python
model_a = SimpleModel(784, 10)
model_b = SimpleModel(2048, 1000)

model_a.describe()
print("---")
model_b.describe()
````

### 연습 문제 3

다음 조건을 만족하는 `TrainingConfig` 클래스를 작성하세요.

- `learning_rate`
- `epochs`
- 두 값을 출력하는 `show()` 메서드


````python
class TrainingConfig:
    def __init__(self, learning_rate, epochs):
        self.learning_rate = learning_rate
        self.epochs = epochs

    def show(self):
        print(f"학습률: {self.learning_rate}")
        print(f"에폭: {self.epochs}")


config = TrainingConfig(learning_rate=0.001, epochs=10)
config.show()
````

## 5. NumPy 배열과 `shape`

PyTorch Tensor를 배우기 전에 NumPy 배열로 다차원 데이터의 형태를 익혀 봅니다.


````python
import numpy as np

vector = np.array([1, 2, 3])
matrix = np.array(
    [
        [1, 2, 3],
        [4, 5, 6],
    ]
)

print("vector:", vector)
print("vector.shape:", vector.shape)

print("matrix:")
print(matrix)
print("matrix.shape:", matrix.shape)
````

### 이미지 배치의 shape

딥러닝 이미지 텐서는 보통 다음 순서입니다.

$$
(N, C, H, W)
$$

- $N$: 배치 크기
- $C$: 채널 수
- $H$: 높이
- $W$: 너비


````python
images = np.zeros(
    (32, 3, 224, 224)
)  # 입력받은 shape 모양대로 모든 원소가 0인 NumPy 배열을 생성하는 함수아다.

print(images.shape)
print("배치 크기:", images.shape[0])
print("채널 수:", images.shape[1])
print("높이:", images.shape[2])
print("너비:", images.shape[3])
````

## 6. `reshape`와 `-1`

배열의 원소 개수는 유지하면서 형태만 바꿉니다.


````python
x = np.arange(24)

print("변경 전:")
print(x)
print(x.shape)

x = x.reshape(
    2, 3, 4
)  # 크기가 3×4인 2차원 배열이 첫번째 축을 따라 2개 쌓인 3차원 배열을 의미

print("\n변경 후:")
print(x)
print(x.shape)
````

### `-1`의 의미

**`-1`은 해당 차원의 크기를 Python/NumPy/PyTorch가 자동으로 계산하라는 뜻입니다.**


````python
x = np.zeros((2, 3, 4))

print("원래 shape:", x.shape)

flattened = x.reshape(2, -1)

print("변경된 shape:", flattened.shape)
````

위 배열은 원래 `(2, 3, 4)`입니다.

첫 번째 차원 `2`는 유지하고 나머지 차원을 하나로 합치면:

$$
3 \times 4 = 12
$$

따라서 결과는 `(2, 12)`가 됩니다.

PyTorch에서 자주 보는 다음 코드도 같은 의미입니다.

```python
x = x.view(x.shape[0], -1)
```

즉, **배치 차원은 유지하고 나머지 차원을 모두 펼치는 것**입니다.

## 9. 최종 확인 문제

다음 코드의 출력 shape를 실행 전에 예상해 보세요.


````python
import numpy as np


class FlattenBatch:
    def __init__(self):
        self.name = "flatten"

    def forward(self, x):
        return x.reshape(x.shape[0], -1)


layer = FlattenBatch()
x = np.zeros((8, 3, 32, 32))
output = layer.forward(x)

print("입력 shape:", x.shape)  # (8, 3, 32, 32)
print("출력 shape:", output.shape)  # (8, 3072)
````
